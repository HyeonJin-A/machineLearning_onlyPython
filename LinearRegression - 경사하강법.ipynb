직선 y=wx+b를 데이터에 가장 근접하게 (cost함수인 mse가 최소가 되도록) 만들어야 함

이 때 경사하강법을 사용함) 미분을 통해서 gradient를 구한 뒤 기존의 w에 뺄셈해주면서 업데이트해 나가는 방법

## 파이썬 만으로 gradient 계산 (cost함수=MSE)

1. 입력 데이터와의 거리 계산
2. cost 함수(MSE) 계산
3. gradient 계산 (cost 함수 미분)
4. w -= gradient
5. 1번으로 루프
6. gradient=0일 때까지 반복

```python
def MSE(x,y,w,b):
    mse = 0
    for xi,yi in zip(x,y):
        mse += pow(w*xi+b-yi,2)/len(x)
    return mse

def calc_gradient_w(x,y,w,b):
    gradient = 0
    for xi,yi in zip(x,y):
        gradient += (2*xi*(w*xi+b-yi))/len(x)
    return gradient

def calc_gradient_b(x,y,w,b):
    gradient = 0
    for xi,yi in zip(x,y):
        gradient += (2*(w*xi+b-yi))/len(x)
    return gradient
```

```python
x = [1.0,2.0,3.0]
y = [3.0,5.0,7.0]
w = 1.0
b = 0.0

LEARNING_RATE = 1.5e-1
w -= calc_gradient_w(x,y,w,b)*LEARNING_RATE
b -= calc_gradient_b(x,y,w,b)*LEARNING_RATE
w,b
```
