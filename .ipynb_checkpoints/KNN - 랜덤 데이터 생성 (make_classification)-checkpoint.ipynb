##  make_classification
https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_classification.html

### 매개변수
- n_samples : 표본 데이터의 수 (default=100)
- n_features : 독립 변수의 수 (defualt=20)
- n_informative : 상관 관계가 있는 변수의 개수 (default=20) 변수 개수와 동일하게 해야할 듯
- n_redundant : 다른 독립 변수와 선형 조합으로 나타나는 변수의 개수 (default=2)
- n_reapeated : 중복된 변수의 개수 (default=0)
- n_classes : 종속 변수의 개수 (default=2)  독립 변수보다 샘플이 많이 섞여 있음
- n_clusters_per_class : 클래스 당 군집의 개수 (default=2)
- weights : 각 클래스에 할당된 표본 수 (default=None)
- flip_y : 임의로 클래스 간 샘플을 섞음. 노이즈를 주어 분류를 어렵게 만듬. (default=0.01) 비율인 듯

```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
SEED = 111
```

```python
# 독립 변수 2개, 샘플50개로 생성
X, Y = make_classification(n_samples=50, n_features=2, n_informative=2, n_redundant=0, random_state=SEED)

plt.scatter(X[:,0], X[:,1], marker='o', c=Y, s=100) #컬러 기준은 Y, 마커사이즈는 100
print(f"X.shape:{X.shape}, Y.shape:{Y.shape}")
```

### 랜덤 입력 데이터 생성

```python
np.random.seed(SEED)
input_data = np.random.randn(1,2)
input_data
```

```python
plt.scatter(X[:,0], X[:,1], marker='o', c=Y, s=100)
plt.scatter(input_data[:,0], input_data[:,1], marker='x', c='red', s=200)
```

## KNN 알고리즘으로 분류

```python
from sklearn.neighbors import KNeighborsClassifier

K = 3
knn = KNeighborsClassifier(n_neighbors=K)  #knn 객체 생성
knn.fit(X, Y)                              #학습
pred = knn.predict(input_data)             #예측
pred
```

```python
plt.scatter(X[:,0], X[:,1], marker='o', c=Y, s=100)
plt.scatter(input_data[:,0], input_data[:,1], marker='x', c=pred, s=200) #컬러를 예측 결과로 설정
```

# K 값을 바꿔가며 결과 확인해보기

샘플이 많이 뒤섞이도록 flip_y값 조절

```python
X, Y = make_classification(n_samples=50, n_features=2, n_informative=2, n_redundant=0, flip_y=0.9, random_state=SEED)

plt.scatter(X[:,0], X[:,1], marker='o', c=Y, s=100)
plt.scatter(input_data[:,0], input_data[:,1], marker='x', c='red', s=200)
print(f"X.shape:{X.shape}, Y.shape:{Y.shape}")
```

```python
fig, ax = plt.subplots(4,4, figsize=(16,16))

for K in range(3,19):
    knn = KNeighborsClassifier(n_neighbors=K)
    knn.fit(X, Y)
    pred = knn.predict(input_data)
    
    if K<7:
        ax[0, K-3].scatter(X[:,0], X[:,1], marker='o', c=Y, s=100)
        ax[0, K-3].scatter(input_data[:,0], input_data[:,1], marker='x', c=pred, s=200)
        ax[0, K-3].set_title(f"K={K}")
    elif K<11:
        ax[1, K-7].scatter(X[:,0], X[:,1], marker='o', c=Y, s=100)
        ax[1, K-7].scatter(input_data[:,0], input_data[:,1], marker='x', c=pred, s=200)
        ax[1, K-7].set_title(f"K={K}")
    elif K<15:
        ax[2, K-11].scatter(X[:,0], X[:,1], marker='o', c=Y, s=100)
        ax[2, K-11].scatter(input_data[:,0], input_data[:,1], marker='x', c=pred, s=200)
        ax[2, K-11].set_title(f"K={K}")
    else:
        ax[3, K-15].scatter(X[:,0], X[:,1], marker='o', c=Y, s=100)
        ax[3, K-15].scatter(input_data[:,0], input_data[:,1], marker='x', c=pred, s=200)
        ax[3, K-15].set_title(f"K={K}")
        
#plt.tight_layout()
plt.show()
```
